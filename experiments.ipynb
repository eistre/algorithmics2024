## Algorithmics group project

A travel route optimization system designed to help travelers plan their journeys efficiently by finding optimal routes between multiple destinations. The system takes into account various factors including:

- Distance between locations
- Travel time
- Associated costs

This implementation leverages graph algorithms to determine the most efficient paths while allowing users to:
- Input custom destinations
- Configure start/end points
- Optimize based on multiple criteria

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import os
import sys
import time
from lib.planner import TravelPlanner, Criteria, CriteriaWeight

sys.path.append(os.getcwd())

# Initialize planner and criteria weights
planner = TravelPlanner(load_from_file=True)
criteria_weights = [CriteriaWeight(Criteria.DISTANCE, 1), 
                   CriteriaWeight(Criteria.DURATION, 0), 
                   CriteriaWeight(Criteria.COST, 0)]

# Define common parameters
start_city = 'Oslo'
end_city = 'Barcelona'
destinations = ['Stockholm', 'Helsinki', 'Barcelona', 'Berlin', 'Vilnius', 
               'Tallinn', 'Copenhagen', 'Frankfurt', 'Warsaw', 'Rome', 'London','Sofia']

# Test different algorithms
algorithms = {
    'Brute Force': planner.optimize_route_brute_force,
    'Greedy': planner.optimize_route_greedy,
    'Simulated Annealing': planner.optimize_route_simulated_annealing,
    'Floyd-Warshall': planner.optimize_route_floyd_warshall,
    'Dijkstra': planner.optimize_route_dijkstra,
    'A*': planner.optimize_route_a_star
}

# Load coordinates
city_coordinates = {}
with open("sample_data/cities_coordinates.txt", "r") as file:
    for line in file:
        city, lat, lon = line.strip().split(",")
        city_coordinates[city] = (float(lat), float(lon))

def plot_routes(routes_data):
    num_routes = len(routes_data)
    cols = 2  # Number of columns in the grid
    rows = (num_routes + cols - 1) // cols  # Calculate required rows
    
    fig = plt.figure(figsize=(20, 10 * rows))
    
    for idx, (name, route_info) in enumerate(routes_data.items()):
        route = route_info['route']
        duration = route_info['duration']
        cost = route_info['cost']
        time_elapsed = route_info['time']
        distance = route_info['distance']
        
        ax = plt.subplot(rows, cols, idx + 1)
        
        # Create a map for each subplot
        m = Basemap(projection='merc', llcrnrlat=35, urcrnrlat=65, 
                   llcrnrlon=-10, urcrnrlon=30, resolution='i', ax=ax)
        
        # Draw map features
        m.drawcountries(linewidth=0.1, color="#7F7F7F")
        m.fillcontinents(color='#F4F5FA', alpha=0.7, lake_color='#D6D6D6')
        m.drawcoastlines(linewidth=0.1, color="#7F7F7F")
        #m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
        m.drawmapboundary(fill_color='#B7C2E4', linewidth=0)
        
        # Plot route
        for i in range(len(route) - 1):
            start_city = route[i]
            end_city = route[i + 1]
            
            if start_city in city_coordinates and end_city in city_coordinates:
                lat1, lon1 = city_coordinates[start_city]
                lat2, lon2 = city_coordinates[end_city]
                
                m.drawgreatcircle(lon1, lat1, lon2, lat2, linewidth=2, color='#5467FA')
                
                x1, y1 = m(lon1, lat1)
                x2, y2 = m(lon2, lat2)
                
                plt.plot(x1, y1, 'o',color='#5467FA', markersize=5)
                plt.text(x1, y1, start_city, fontsize=8, ha='right', va='bottom', color='#2C2C2C')
                
                if i == len(route) - 2:  # Plot the last city
                    plt.plot(x2, y2, 'o',color='#5467FA', markersize=5)
                    plt.text(x2, y2, end_city, fontsize=8, ha='right', va='bottom',color='#2C2C2C')
        
        # Add title and metrics
        plt.title(f"{name}\n Distance: {distance:.2f} | Duration: {duration:.1f}h | Cost: {cost:.1f}€ | Time: {time_elapsed:.5f}s",
                 fontsize=12, pad=10)
    
    plt.tight_layout(pad=3.0)
    plt.show()

# Store results and run algorithms
routes_data = {}
for name, algorithm in algorithms.items():
    print(f'Running {name} algorithm...')
    
    start_time = time.time()
    optimized_route, distance, duration, cost = algorithm(
        start=start_city, 
        end=end_city, 
        destinations=destinations, 
        criteria_weights=criteria_weights
    )
    end_time = time.time()
    
    routes_data[name] = {
        'route': optimized_route,
        'distance': distance,
        'duration': duration,
        'cost': cost,
        'time': end_time - start_time
    }
    
    print(f'Optimized route: {" -> ".join(optimized_route)}')
    print(f'Distance: {distance:.2f} km')
    print(f'Duration: {duration:.2f} hours')
    print(f'Cost: {cost:.2f} EUR')
    print(f'Time elapsed: {end_time - start_time:.2f} seconds\n')

# Plot all routes in a grid
plot_routes(routes_data)
```

## Running time

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

def plot_algorithm_performance_with_custom_zoom(df):
    # Create the figure and axis
    plt.figure(figsize=(12, 6))
    
    # Define a color map for consistent coloring
    algorithms = df['Algorithm'].unique()
    colors = plt.cm.get_cmap('tab10', len(algorithms))  # Change 'tab10' to any colormap you prefer
    
    # Plot a line for each algorithm
    for i, algorithm in enumerate(algorithms):
        algorithm_data = df[df['Algorithm'] == algorithm]
        plt.plot(algorithm_data['n_cities'], 
                 algorithm_data['Running_time'], 
                 marker='o', 
                 label=algorithm, 
                 color=colors(i))  # Assign consistent color to each algorithm
    
    # Main plot customization
    plt.title('Algorithm Running Time vs Number of Cities', fontsize=14, pad=15)
    plt.xlabel('Number of Cities (including start & end)', fontsize=12)
    plt.ylabel('Running Time (seconds)', fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Add the inset (zoomed) plot
    ax_inset = inset_axes(plt.gca(),  # Parent axes
                          width="20%",  # Width of inset (percentage of parent)
                          height="20%",  # Height of inset (percentage of parent)
                          loc='upper center')  # Location of inset
    
    # Define the zoom region: you can adjust these values based on your needs
    zoom_x_min, zoom_x_max = 8, 12  # Define range of 'n_cities' you want to zoom into
    zoom_y_min, zoom_y_max = 0, 0.04  # Define range of 'Running_time' you want to zoom into
    
    # Zoomed region - filter the data
    zoom_region = df[(df['n_cities'] >= zoom_x_min) & (df['n_cities'] <= zoom_x_max) &
                     (df['Running_time'] >= zoom_y_min) & (df['Running_time'] <= zoom_y_max)]
    
    # Plot the zoomed region with the same colors
    for i, algorithm in enumerate(algorithms):
        algorithm_data = zoom_region[zoom_region['Algorithm'] == algorithm]
        ax_inset.plot(algorithm_data['n_cities'], 
                      algorithm_data['Running_time'], 
                      marker='o', 
                      label=algorithm, 
                      color=colors(i))  # Consistent color
    
    # Set axis limits for the zoomed-in plot
    ax_inset.set_xlim(zoom_x_min, zoom_x_max)  # Zoom in on x-axis
    ax_inset.set_ylim(zoom_y_min, zoom_y_max)  # Zoom in on y-axis
    ax_inset.set_xticks(range(zoom_x_min, zoom_x_max + 1))
    
    # Customize the inset
    ax_inset.set_title("Zoomed In", fontsize=10)
    ax_inset.set_xlabel("n_cities", fontsize=8)
    ax_inset.set_ylabel("Time (s)", fontsize=8)
    ax_inset.tick_params(axis='both', which='major', labelsize=8)
    ax_inset.grid(True, linestyle='--', alpha=0.7)
    
    # Optional: Add bounding box on the main plot showing zoom region
    plt.gca().add_patch(plt.Rectangle((zoom_x_min, zoom_y_min), 
                                      zoom_x_max - zoom_x_min, 
                                      zoom_y_max - zoom_y_min,
                                      edgecolor='red', facecolor='none', 
                                      linestyle='--', linewidth=1.5))
    
    return plt

# Create and display the plot
plot = plot_algorithm_performance_with_custom_zoom(df)
plot.show()

```

```python
import pandas as pd
import numpy as np

# Initialize the new df_overview DataFrame
columns = ["Algorithm", "Dist. Diff. (w/ End) n=12", "Dist. Diff. (w/o End) n=12", "Avg Running Time", "Min Running time", "Max Running time"]
df_overview = pd.DataFrame(columns=columns)

# Get the unique algorithms
algorithms = df2["Algorithm"].unique()

df_w_end_n12 = df2[(df2["End_city"] == True) & (df2["n_cities"] == 12)]
dist_min = df_w_end_n12["Distance"].min()

df_wo_end_n12 = df2[(df2["End_city"] == False) & (df2["n_cities"] == 12)]
dist_min_wo_end = df_wo_end_n12["Distance"].min()
print(dist_min_wo_end)

for algorithm in algorithms:
    # Filter df2 by the algorithm
    algorithm_df = df2[df2["Algorithm"] == algorithm]
    
    # Filter by n=12
    df_n_12 = algorithm_df[algorithm_df["n_cities"] == 12]
    
    # Dist. Diff. (w/ End) and Dist. Diff. (w/o End)
    dist_with_end_min = df_n_12[df_n_12["End_city"] == True]["Distance"].min()  # Min distance with End
    
    dist_without_end = df_n_12[df_n_12["End_city"] == False]["Distance"].min()  # Min distance without End
    
    # If both distances are available, calculate difference
    dist_diff_with_end = dist_min - dist_with_end_min
    dist_diff_wo_end = dist_min_wo_end - dist_without_end
    
    # Avg Running Time
    avg_running_time = algorithm_df["Running_time"].mean()

    min_running_time = df_n_12["Running_time"].min() 
    max_running_time = df_n_12["Running_time"].max()
    
    # Min and Max Running Time with corresponding n
    min_running_time_row = df_n_12[df_n_12["Running_time"] == min_running_time].iloc[0]
    max_running_time_row = df_n_12[df_n_12["Running_time"] == max_running_time].iloc[0]
    
    min_n = min_running_time_row["n_cities"]
    max_n = max_running_time_row["n_cities"]
    
    # Format for "Min Running time" and "Max Running time"
    min_running_time_str = f"{round(min_running_time, 4)} s, n={min_n}, end_city = {min_running_time_row["End_city"]}"
    max_running_time_str = f"{round(max_running_time, 4)} s, n={max_n}, end_city = {max_running_time_row["End_city"]}"
    
    # Create the new row for df_overview
    new_row = {
        "Algorithm": algorithm,
        "Dist. Diff. (w/ End) n=12": f"{abs(round(dist_diff_with_end, 2))} km",
        "Dist. Diff. (w/o End) n=12": f"{abs(round(dist_diff_wo_end, 2))} km",
        "Avg Running Time": f"{round(avg_running_time, 4)} s",
        "Min Running time": min_running_time_str,
        "Max Running time": max_running_time_str
    }
    
    # Append the new row to df_overview
    df_overview = pd.concat([df_overview, pd.DataFrame([new_row])], ignore_index=True)

df_overview

```
